```python
import pandas as pd
import numpy as np
```

```python
s = pd.Series([11,2,3,4,5])
print(s)
 #索引可以自己控制
s1 = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e'],name = '鱼粉') 
print(s1)
```

```python
s1 = pd.Series([1,2,3,4,5],index = ['a','b','c','d','e']) 
print(s.index)
print(s1.index)
print(s1.values)    #键值
print(s1.dtype)
print(s1.loc['a'])   #显示索引：查找索引'a’的值
print(s1.iloc[1])    #隐式索引：查找索引的值
print(s1.at['a'])    #显示索引        不能切片
print(s1.iat[3])     #隐式索引(位置)        不能切片
```

```python
#访问数据
s3 = pd.Series([4,5,75,525,356346],index = ['a','b','c','d','e'],name = '一')
print(s3['a'])
print(s3[s3<55])
print('头',s3.head(2))   #查看前n行数据，默认五行
print(s3.tail(3))        #查看后n行数据，默认五行
```

```python
#常见函数
s4 = pd.Series([10,2,np.nan,None,2,5,6],
               index = ['a','b','c','d','e','f','g'],name ='数据')
print(s4)
```

```python
s4.describe()
```

```python
print(s4.head(3))
print(s4.tail(5))
```

```python
#获取索引
print(s4.keys())     #方法
print(s4.index)      #属性
```

```python
print(s4.isna())  #检查每一个元素是否为确实值
print(s4.isin([5,4]))   #检查数据是否在参数集合里
```

```python
print(s4.sort_values())    #排序
print(s.quantile(0.25))    #中位数
print(s4.mode())    #众数
print(s4.value_counts())   #每个元素的计数

```

```python
print(s4.drop_duplicates())   #去重
print(s4.unique())            #去重  包含nan
print(s4.nunique())           #去重后的元素个数  不包含nun
```
