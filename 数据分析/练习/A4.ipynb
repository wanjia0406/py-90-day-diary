```python
import pandas as pd
import numpy as np
```

```python
'''创建一个包含10名学生数学成绩的Series，
成绩范围在50-100之间。然后计算平均分、最高分、最低分，
并找出高于平均分的学生人数。
'''
np.random.seed(42)
score = pd.Series(np.random.randint(50,101,10),
                  index = ['学生'+str(i)for i in range(1,11)])
print(score)
ping = score.mean()
print("平均分:",ping)
print("最高分:",score.max())
print("最低分:",score.min())
print("高于平均分:",np.sum(score>ping))
print("高于平均分:",(score>ping).sum())
print("高于平均分:",np.sum(score>ping))
print("高于平均分:",score[score>ping].count())
print("高于平均分:",np.sum(score>ping))
print("高于平均分:",score.where(score>ping).dropna().count())
```

```python
'''给定某城市一周每天的最高温度Series，完成以下任务：
找出温度超过30度的天数
计算平均温度
将温度从高到低排序
找出温度变化最大的两天'''
temperatures = pd.Series([28, 31, 29, 32, 30, 27, 33],
                         index=['周一', '周二', '周三', '周四', '周五', '周六', '周日'])
print("温度超过30的天数:",temperatures[temperatures>30].count())
print("平均温度:",temperatures.mean())
print("从高到低排序:",temperatures.sort_values(ascending=False))
# temperatures.diff()         #计算变化值
t = temperatures.diff().abs().idxmax()
t1 = temperatures.diff().abs().max()
# t.sort_values(ascending=False).keys()
print("温差最大的:"f"{t} 与前一天，温差 {t1}℃")
```

```python
'''给定某股票连续10个交易日的收盘价Series：
计算每日收益率（当日收盘价/前日收盘价－1）
找出收益率最高和最低的日期
计算波动率（收益率的标准差）'''
prices = pd.Series([102.3, 103.5, 105.1, 104.8, 106.2, 107.0, 106.5, 108.1, 109.3, 110.2],
                   index=pd.date_range('2023-01-01', periods=10))
# print(prices)
profit = prices.pct_change().dropna()
print(profit)
max_date = profit.idxmax()
print("最大日期:",max_date)
min_date = profit.idxmin()
print("最低日期:",min_date)
print("波动率:",profit.std())
```

```python
'''某产品过去12个月的销售量Series:
计算季度平均销量(每3个月为一个季度)
找出销量最高的月份
计算月环比增长率
找出连续增长超过2个月的月份'''

sales = pd.Series([120, 135, 145, 160, 155, 170, 180, 175, 190, 200, 210, 220],
                  index=pd.date_range('2022-01-01', periods=12, freq='M'))
# print(sales)
print("销量最高:",sales.idxmax())
quarter_mean = sales.resample('QS').mean()
print("季度平均销量：")
print(quarter_mean)
mom = sales.pct_change().dropna()
print("月环比增长率：")
print(mom, '\n')
# 5. 无循环版：连续增长超过 2 个月的终点月
growing = (mom > 0).astype(int)                 # True→1, False→0
roll_sum = growing.rolling(window=3).sum()      # 连续3个月之和
# 和为3 ⇒ 3个月都增长，取这些月份的索引
long_growth = roll_sum[roll_sum == 3].index
print("连续增长超过2个月的月份（终点月）：")
print(long_growth.strftime('%Y-%m').tolist())
```

```python
'''某商店每小时销售额Series：
按天重采样计算每日总销售额
计算每天营业时间（8:00-22:00）和非营业时间的销售额比例
找出销售额最高的3个小时'''

np.random.seed(42)
hourly_sales = pd.Series(np.random.randint(0, 100, 24),
                         index=pd.date_range('2025-01-01', periods=24, freq='H'))
a = hourly_sales.resample('D').sum()
print(a)
in_hours  = hourly_sales.between_time('08:00', '22:00').sum()
out_hours = hourly_sales.sum() - in_hours
ratio_in  = in_hours  / (in_hours + out_hours)
ratio_out = out_hours / (in_hours + out_hours)
print('营业时间销售额  ：', in_hours,  f'元  占比：{ratio_in:.2%}')
print('非营业时间销售额：', out_hours, f'元  占比：{ratio_out:.2%}\n')
top3 = hourly_sales.nlargest(3)
print('销售额最高的 3 个小时：')
for ts, val in top3.items():
    print(ts.strftime('%Y-%m-%d %H:%M'), val, '元')
```
