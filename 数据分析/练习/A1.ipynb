##ndarray的创建

```python
#等差数列
arr = np.arange(1,11,2)
print(arr)
```

```python
# 等间隔数列
arr = np.linspace(1,10,5)  # 开始，结束，多少份（平均）
print(arr)
```

```python
#对数间隔数列
arr = np.logspace(1,4,2,base = 2)
print(arr)
```

```python
arr = np.logspace(1,4,2,base=2)
```

```python
arr = np.eye(3)   #单位矩阵
arr1 = np.diag([1,2,3])    #对角矩阵
print(arr1)
```

```python
arr = np.random.rand(2,3)  #生成0,1之间的随机浮点数
arr1 = np.random.uniform(3,6,(2,3))  #生成指点范围（3,6）之间的随机浮点数  
```

```python
arr = np.array([1, 4, 9], dtype=int)
result = np.sqrt(arr)
print(result)
print(np.sqrt([1,4,9]).astype(int))
```

```python
arr = np.array([1,2,3])
print(np.mean(arr))   #平均值
print(np.cumsum(arr))  #累积和
print(np.cumprod(arr))   # 累积积
print(np.median([2,8,3]))   #中位数
print(np.percentile(arr,50))  #分位数
print(np.var([1,2,3]))    #方差
print(np.std([1,2,3]))     #标准差
```

比较函数

```python
print(np.greater([1,2,3,4,5,6],4))   #是否大于
print(np.less([1,2,3,4,5,6],4))     #是否小于
print(np.equal([1,2,3,4,5,6],4))     #是否等于
```

```python
arr = np.array([1,2,3,4,5,6,7])
print(np.where(arr>3,1,2))     #三目运算符
```

```python
score = np.random.randint(1,100,10)
print(score)
print(np.where(
    score<=60,'不及格',np.where(
        score<=80,'良好','优秀')
))
```

```python
print(np.select([score<60,(score>=60) & (score<80),(score>=80) &(score<=100)],['不及格','良好','优秀'],default='未知'))
```

```python
np.random.seed(0)
arr2 = np.random.randint(1,101,10)
print(arr2)
print(np.sort(arr2))     #排序
print(np.argsort(arr2))  #数值相关索引的排序
```
