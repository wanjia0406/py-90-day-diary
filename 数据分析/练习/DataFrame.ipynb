```python
import numpy as np
import pandas as pd
```

```python
s1 = pd.Series([1,2,3,4,5])
s2 = pd.Series([11,12,13,14,15])
s = pd.DataFrame({"第一列:":s1,"第二列":s2})
df = pd.DataFrame({
    "id":[1,2,3,4,5],
    "name":['a','s','d','f','g'],
    "age":[15,13,14,12,13],
    "score":[88,80,75,45,79]
},index = [1,2,3,4,5],
columns=['name','score','age','id']  #调换列的位置顺序  
)
df
```

```python
print("行索引:",df.index)
print("列标签:",df.columns)
print(df.dtypes)
print('维度:',df.ndim)
print('形状:',df.shape)
print('元素个数:',df.size)
print('行列转置',df.T)
```

```python
#获取元素   loc iloc at iat
# print(df)
print(df.loc[4])
print(df.iloc[3])
#单个元素
print(df.at[3,'score'])
print(df.iat[2,1])
print(df.loc[3,'score'])
```
