###折线图

```python
#绘制折线图
import matplotlib.pyplot as plt
from matplotlib import rcParams   #字体  作用：一次性全局改图样
rcParams['font.family'] = 'SimHei'

#创建图表，设置大小
plt.figure(figsize=(6,3))

#绘制图的数据
month = ['1月','2月','3月','4月','5月']
sales = [169,65,213,155,130]
plt.xticks(rotation = 30,fontsize=12)   # 角度，字体大小
plt.yticks(rotation = 30,fontsize=12)   # 角度，字体大小

#绘制折线图
plt.plot(month, sales,          # 横轴 month，纵轴 sales
         label='产品A',          # 图例文字
         color='orange',         # 折线颜色：橙色
         linewidth=2,            # 线宽 2 磅
         linestyle='--',         # 线型：虚线
         marker='o'              # 数据点标记：实心圆
        )
plt.title('2025销售趋势',color ='red',fontsize = 20)    #添加标题
plt.xlabel('月份',fontsize = 20)             #添加坐标轴的标签
plt.ylabel('销售额(万元)',fontsize = 20)      #添加坐标轴的标签
plt.legend(loc = 'upper left')
# plt.grid(axis = 'y')     #添加网格线 (True)/(axis ='y')/(axis = 'x')
plt.grid(axis = 'y',alpha = 0.2,color = 'blue',linestyle = '--')
plt.ylim(0,250)             #设置y轴的范围

#在每个数据点上显示数值
for x,y in zip(month,sales):
     plt.text(x,y+4,str(y),ha='center',va='bottom',fontsize = 15)

#显示图表
plt.show()
```

###柱状图

```python
#绘柱状图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'

#条形图尺寸
plt.figure(figsize=(10,6))

#绘制图的数据
subjects = ['语文','数学','英语','科学']
scores = [89,98,78,91]

#绘制柱状图
bars = plt.bar(subjects,scores,
        label = '小明',
        color='orange',         # 折线颜色：橙色
         linewidth=2,            # 线宽 2 磅
         linestyle='--',         # 线型：虚线
         width = 0.5
        )
plt.title('学生各科成绩',color = 'gold',fontsize = 40)     #标题
plt.xlabel('科目',fontsize = 20,color = 'gold')       #添加坐标轴的标签
plt.ylabel('分数',fontsize = 20,color = 'gold')       #添加坐标轴的标签
plt.xticks(rotation = 30,fontsize = 15)
plt.yticks(rotation = 30,fontsize = 15)
plt.legend(loc ='upper right')
plt.grid(axis ='y',color = 'blue',alpha = 0.4,linestyle = '--')
plt.ylim(50,120)                 #设置y轴的范围
#在每个数据点上显示数值
# for x,y in zip(subjects,scores):
#     plt.text(x,y+3,str(y),ha ='left',va = 'bottom',fontsize = 20)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2, height + 3, 
             str(height), ha='center', va='bottom', fontsize=20) 
plt.tight_layout()       #自动优化排版


#显示图表
plt.show()
```

###条形图

```python
#绘制条形图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'

#条形图尺寸
plt.figure(figsize=(10,6))

#绘制图的数据
county = ['United States','China','Japan','Germany','India']
GDP = [89,98,78,61,45]

# 按GDP从小到大排序
sorted_data = sorted(zip(GDP, county))
sorted_GDP = [item[0] for item in sorted_data]
sorted_county = [item[1] for item in sorted_data]

#绘制柱状图
bars =plt.barh(sorted_county,sorted_GDP,
        label = 'GDP',
        color = 'orange'
        )
plt.title('全球各国薪资水平',color = 'gold',fontsize = 40)
plt.xlabel('GDP',fontsize = 40,color = 'gold')   #绘制标签
plt.ylabel('county',fontsize = 40,color ='gold') #绘制标签
plt.xticks(rotation = 30,fontsize = 20)
plt.yticks(rotation = 30,fontsize = 10)
plt.legend(loc = 'upper right')
plt.grid(axis ='x',color= 'black',alpha = 0.3,linestyle = '--')
sorted_data = sorted(zip(GDP, county))  # 去掉reverse=True
plt.tight_layout()   #自动优化排版
# 在柱子右侧标数值
for bar, value in zip(bars,sorted_GDP):
    plt.text(bar.get_width() + 1,            # x：柱宽再右移 1
             bar.get_y() + bar.get_height()/2,  # y：柱中心
             str(value), va='center', ha='left', fontsize=15)


#显示图表
plt.show()
```

###饼图

```python
#绘制饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'

#条形图尺寸
plt.figure(figsize=(10,6))

#绘制图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
color=['#ff33bb','#99ff99','gold','#ff4499','#66b3ff']                       #配图
 
#绘制柱状图
plt.pie(times,labels=things,
        autopct='%.2f%%',          #显示百分比
        startangle=90,
        colors = color              #调整初始化图的角度
        )
plt.title('时间分布表',color = 'gold',fontsize = 30)

plt.tight_layout()

#显示图表
plt.show()
```

```python
#绘制环形图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'

#条形图尺寸
plt.figure(figsize=(10,6))

#绘制图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
color=['#ff33bb','#99ff99','gold','#ff4499','#66b3ff']                        #配图
 
#绘制柱状图
plt.pie(times,labels=things,
        autopct='%.2f%%',          #显示百分比
        startangle=90,              #调整初始化图的角度
        colors = color,             #圆环的配色
        wedgeprops={'width':0.7},   #设置圆环的宽度
        pctdistance=0.8             #设置百分比的位置
        )
                   
plt.title('时间分布表',color = 'gold',fontsize = 30)
plt.text(0,0,'百分比:100%')
plt.tight_layout()

#显示图表
plt.show()
```

```python
#绘制爆炸式饼图
import matplotlib.pyplot as plt
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'

#条形图尺寸
plt.figure(figsize=(10,6))

#绘制图的数据
things = ['学习','娱乐','运动','睡觉','其他']
times = [6,4,1,8,5]
color=['#ff33bb','#99ff99','gold','#ff4499','#66b3ff']                       #配图
explode =[0.1,0,0,0,0]   #设置突出块的位置

#绘制柱状图
plt.pie(times,labels=things,
        autopct='%.2f%%',          #显示百分比
        startangle=90,
        colors = color,              #调整初始化图的角度
        explode=explode,
        shadow = True
        )
plt.title('时间分布表',color = 'gold',fontsize = 30)

plt.tight_layout()

#显示图表
plt.show()
```

###散点图

```python
#绘制散点图
import matplotlib.pyplot as plt
import random
from matplotlib import rcParams
rcParams['font.sans-serif'] = 'SimHei'

#条形图尺寸
plt.figure(figsize=(10,6))

#绘制图的数据
x = []
y = []
for i in range(1000):
    tem = random.uniform(0,10)
    x.append(tem)
    y.append(tem*2+random.gauss(0,2))

#绘制柱状图
plt.scatter(x, y,           # 必需参数：x坐标和y坐标的数据
            color = 'blue',  # 散点的颜色：蓝色
            alpha = 0.5,     # 透明度：50% 半透明
            s = 20,          # 散点的大小：20平方磅
            label ='数据'
            )
plt.title('变量x,yy的关系',color = 'gold',fontsize = 30)
plt.xlabel('x自变量',fontsize = 20)             #添加坐标轴的标签
plt.ylabel('y因变量',fontsize = 20)      #添加坐标轴的标签
plt.xticks(rotation = 30,fontsize = 20)
plt.yticks(rotation = 30,fontsize = 20)
plt.legend(loc = 'upper left')
# plt.grid(axis = 'y')     #添加网格线 (True)/(axis ='y')/(axis = 'x')
plt.grid(True,alpha = 0.2,color = 'blue',linestyle = '--')  #网格线
plt.ylim(0,30)
plt.plot([0,10],[0,20],color = 'red',linewidth =2,linestyle = '--')

#显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt           # 导入绘图库
import numpy as np                        # 导入数值计算库

# 1. 数据
month  = np.arange(1, 13)                 # 生成 1~12 月份数组
sales  = np.array([215,232,258,281,310,335,362,378,395,420,450,480])  # 销售额（万元）
profit = np.array([15,18,22,25,30,35,38,42,45,48,52,58])              # 利润率（%）

# 2. 创建画布和左侧 Y 轴坐标系
fig, ax_bar = plt.subplots(figsize=(10,5))  # 新建图窗，返回(图窗, 左侧轴)

# 柱形：左侧 Y 轴
bars = ax_bar.bar(month, sales, color='steelblue', alpha=0.7, label='销售额（万元）')  # 画柱形
ax_bar.set_xlabel('月份')                    # 设置 X 轴标题
ax_bar.set_ylabel('销售额（万元）', color='steelblue')  # 左侧 Y 轴标题+颜色
ax_bar.tick_params(axis='y', labelcolor='steelblue')   # 左侧 Y 轴刻度颜色

# 折线：右侧 Y 轴
ax_line = ax_bar.twinx()                     # 复制 X 轴，新建右侧 Y 轴
line = ax_line.plot(month, profit, color='orangered', marker='o', linewidth=2,
                    label='利润率（%）')       # 画折线
ax_line.set_ylabel('利润率（%）', color='orangered')  # 右侧 Y 轴标题+颜色
ax_line.tick_params(axis='y', labelcolor='orangered')  # 右侧 Y 轴刻度颜色

# 3. 辅助元素
ax_bar.set_title('2024 年月度销售额与利润率')  # 图标题
fig.tight_layout()                          # 自动调整子图间距
plt.show()                                  # 弹出窗口显示
```
