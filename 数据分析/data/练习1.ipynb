```python
# 创建示例DataFrame
df = pd.DataFrame({
    'numeric_nan': [1, 2, np.nan, 4],
    'numeric_na': [1, 2, pd.NA, 4],
    'string_nan': ['a', 'b', np.nan, 'd'],
    'string_na': ['a', 'b', pd.NA, 'd'],
    'boolean_na': [True, False, pd.NA, True]
})

print("DataFrame信息:")
print(df.dtypes)
print("\n缺失值检测:")
print(df.isna())
```
